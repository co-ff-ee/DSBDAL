```python
# One Hot Encoding
encoded_df = pd.get_dummies(df1, columns=['Species'], dtype=int)

print((encoded_df).to_string())
```

```python
print(encoded_df.columns)
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)
encoded = encoder.fit_transform(df1[['Species']])

encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(['Species']))

final_df = pd.concat([df1.drop('Species', axis=1), encoded_df], axis=1)

```
